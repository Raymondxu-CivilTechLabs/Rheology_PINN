```python

# ============================================================
# 【PyTorch + DeepXDE 版本】第一步：设置后端并导入依赖
# ============================================================
# 关键一步：必须在 import deepxde 之前设置环境变量，
# 告诉 DeepXDE 使用 PyTorch 而不是默认的 TensorFlow 作为计算后端。
import os
os.environ["DDE_BACKEND"] = "pytorch"

import deepxde as dde          # PINN框架，现在将使用 PyTorch 后端
import torch                   # PyTorch：替代原来的 TensorFlow，用于 PDE 函数中的张量运算
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from scipy.interpolate import griddata
import scipy.io
from PIL import Image
```

```python
#材料参数（论文 Table 1，w/c=0.4 那行，论文公式(6)(7)(8)
def solution(x1,x2):
    y_f = 12.1
    A_0 = 43.6
    
    A_1 =131.2
    
    C = 56.1
    a=2100
    b=9.45
    
    #（最核心）：把公式(6)(7)(8)合三为一
    y = y_f+(A_0+(A_1-A_0)*np.exp(-x2*x1/(a+b*x2)))*(np.arcsinh(x2/C))
    
    return y
```

```python

# ============================================================
# 【PyTorch + DeepXDE 版本】第二步：定义物理方程(PDE)并训练
# ============================================================
# 对应原论文公式 (21)：物理残差 = 解析解(Eq.8) - 网络预测值
# 【改动说明】只改了两行API：
#   tf.exp(...)        → torch.exp(...)
#   tf.math.asinh(...) → torch.asinh(...)
# 其余网络结构、训练参数与原版完全一致，保证可复现性。
def PDE(x, y):
    x1, x2 = x[:, 0:1], x[:, 1:2]   # x1=时间t, x2=剪切速率γ̇

    # 材料参数（论文 Table 1，w/c=0.4）
    y_f = 12.1
    A_0 = 43.6
    A_1 = 131.2
    C   = 56.1
    a   = 2100
    b   = 9.45

    # 物理约束残差：解析解 - 网络输出 = 0
    # 【关键改动】torch.exp 和 torch.asinh 替代了原来的 tf 版本
    # DeepXDE 在 PyTorch 后端下，x 和 y 均为 torch.Tensor，
    # 所以必须用 torch 的算子，而不是 numpy 或 tf。
    eq1 = (y_f
           + (A_0 + (A_1 - A_0) * torch.exp(-x2 * x1 / (a + b * x2)))
           * torch.asinh(x2 / C)
           - y)
    return eq1

def func(x):
    # 解析解，用于生成训练数据的"真实值"标签
    return solution(x[:, 0:1], x[:, 1:2])

# ————————————————————————————————————————
# 几何域定义：t ∈ [0,200], γ̇ ∈ [0,200]
# 与原版完全一致
# ————————————————————————————————————————
geom = dde.geometry.Rectangle([0, 0], [200, 200])

# 数据集：3500个内部配置点 + 1000个测试点，伪随机分布
data = dde.data.PDE(
    geom, PDE, [],
    num_domain=3500,
    num_boundary=1000,
    train_distribution='pseudo',
    solution=func,
    num_test=None
)

# ————————————————————————————————————————
# 网络架构：2输入 → 3层×50节点 → 1输出（与原版一致）
# ————————————————————————————————————————
layer_size   = [2] + [50] * 3 + [1]
activation   = "tanh"          # tanh 激活函数，适合捕捉光滑物理场
initializer  = "Glorot uniform"

net   = dde.maps.FNN(layer_size, activation, initializer)
model = dde.Model(data, net)

# Adam优化器，lr=0.001，训练50000轮（与原版一致）
model.compile("adam", lr=0.001, metrics=["l2 relative error"])
losshistory, train_state = model.train(epochs=50000)

dde.saveplot(losshistory, train_state, issave=True, isplot=True)
```
